Plot a map of broadband bi-hemispherical albedo from Sentinel-3 data using earthkit-plots.

```python
from earthkit import plots as ekp

# select a portion of the data to plot
lat_range = (48, 61)
lon_range = (-12, 3)
time_stamp = "2020-01-01"
plot_data = ds["AL_BH_BB"].sel(time=time_stamp, method="nearest").sel(latitude=slice(*lat_range), longitude=slice(*lon_range)).compute()

chart = ekp.Map()

chart.grid_cells(plot_data)

chart.coastlines()
chart.gridlines()

chart.title("{long_name}")

chart.show()
```

## Advanced usage

If your workflow requires access to larger amounts of data, it is recommended that you include a retry mechanism.
This is not provided in the current default zarr engine for xarray, instead we can define a custom "`store`" which is
used when connecting to the remote zarr data.

```python
!pip install -q obstore
import xarray as xr
from obstore.store import HTTPStore
from zarr.storage import ObjectStore

# Geo-chunked Sentinel-3 data
geochunked_sentinel_3_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/sentinel-3/geoChunked.zarr"

# Time-chunked Sentinel-3 data
timechunked_sentinel_3_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/sentinel-3/timeChunked.zarr"

# Geo-chunked SPOT data
geochunked_spot_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/spot/geoChunked.zarr"

# Time-chunked SPOT data
timechunked_spot_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/spot/timeChunked.zarr"

# Geo-chunked Proba data
geochunked_proba_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/proba/geoChunked.zarr"

# Time-chunked Proba data
timechunked_proba_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/proba/timeChunked.zarr"

# Geo-chunked NOAA-7 data
geochunked_noaa_7_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/noaa-7/geoChunked.zarr"

# Time-chunked NOAA-7 data
timechunked_noaa_7_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/noaa-7/timeChunked.zarr"

# Geo-chunked NOAA-9 data
geochunked_noaa_9_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/noaa-9/geoChunked.zarr"

# Time-chunked NOAA-9 data
timechunked_noaa_9_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/noaa-9/timeChunked.zarr"

# Geo-chunked NOAA-11 data
geochunked_noaa_11_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/noaa-11/geoChunked.zarr"

# Time-chunked NOAA-11 data
timechunked_noaa_11_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/noaa-11/timeChunked.zarr"

# Geo-chunked NOAA-14 data
geochunked_noaa_14_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/noaa-14/geoChunked.zarr"

# Time-chunked NOAA-14 data
timechunked_noaa_14_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/noaa-14/timeChunked.zarr"

# Geo-chunked NOAA-16 data
geochunked_noaa_16_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/noaa-16/geoChunked.zarr"

# Time-chunked NOAA-16 data
timechunked_noaa_16_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/noaa-16/timeChunked.zarr"

# Geo-chunked NOAA-17 data
geochunked_noaa_17_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-050/arco/satellite_albedo/noaa-17/geoChunked.zarr"

# Time-chunked NOAA-17 data
timechunked_noaa_17_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-050/arco/satellite_albedo/noaa-17/timeChunked.zarr"


# Use obstore's HTTPStore to create a store with retry configuration,
# and then wrap it in a zarr ObjectStore to read with xarray.
# See https://github.com/developmentseed/obstore/blob/main/obstore/python/obstore/_store/_retry.pyi
# for more details on the retry configuration options.
http_store = HTTPStore(
    geochunked_sentinel_3_url,
    client_options={
        "default_headers": {"Authorization": f"Bearer {cdsapi_key}"},
    },
)
store = ObjectStore(http_store, read_only=True)
ds = xr.open_zarr(store)
ds
```

The result is the same as the previous example, this just provides a more robust connection, preventing your
workflow crashing due to temporary network issues.

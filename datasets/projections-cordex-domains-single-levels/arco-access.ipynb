## Plug and play access

The code below provides a simple plug and play example of how to use the Zarr Store for light-weight access.

This dataset contains **213 sub-items**, each corresponding to a specific combination of:
- **Domain** (e.g. `EUR-11` for European domain at ~12.5 km resolution)
- **Scenario** (e.g. `rcp45` for RCP 4.5)
- **Driving GCM** (e.g. `mohc_hadgem2_es`)
- **Regional Climate Model** (e.g. `clmcom_clm_cclm4_8_17`)
- **Ensemble member** (e.g. `r1i1p1`)
- **Temporal frequency** (e.g. `day`)

The sub-item identifier follows the pattern `{domain}-{scenario}-{frequency}-{gcm}-{rcm}-{ensemble}`.
The example below uses one specific sub-item; replace the sub-item identifier in the URL to access other combinations.

```python
import xarray as xr

# Geo-chunked data for EUR-11, RCP 4.5, MOHC-HadGEM2-ES, CLMcom-CLM-CCLM4-8-17, r1i1p1, daily
geochunked_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-011/arco/projections_cordex_domains_single_levels/eur11-rcp45-day-mohc_hadgem2_es-clmcom_clm_cclm4_8_17-r1i1p1/geoChunked.zarr"

# Time-chunked data for EUR-11, RCP 4.5, MOHC-HadGEM2-ES, CLMcom-CLM-CCLM4-8-17, r1i1p1, daily
timechunked_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-011/arco/projections_cordex_domains_single_levels/eur11-rcp45-day-mohc_hadgem2_es-clmcom_clm_cclm4_8_17-r1i1p1/timeChunked.zarr"


# Open the zarr store with xarray, users must insert their API key where indicated.
ds = xr.open_zarr(
    geochunked_url,
    consolidated=True,
     storage_options = {
        "headers": {"Authorization": f"Bearer {cdsapi_key}"}
    }
)
 
# Inspect the variables
ds
```

Extract a point and plot a time-series of daily precipitation using earthkit-plots.

```python
from earthkit import plots as ekp

# Select variable to plot
variable_name = "pr"
plot_data = ds[variable_name].sel(
    latitude=48.85, longitude=2.35, method="nearest"
)

chart = ekp.TimeSeries()

chart.line(plot_data)

chart.title("Daily precipitation at Paris")
chart.legend()

chart.show()
```

## Advanced usage

If your workflow requires access to larger amounts of data, it is recommended that you include a retry mechanism.
This is not provided in the current default zarr engine for xarray, instead we can define a custom "`store`" which is
used when connecting to the remote zarr data.

```python
!pip install -q obstore
import xarray as xr
from obstore.store import HTTPStore
from zarr.storage import ObjectStore

# Geo-chunked data for EUR-11, RCP 4.5, MOHC-HadGEM2-ES, CLMcom-CLM-CCLM4-8-17, r1i1p1, daily
geochunked_url = "https://arco.datastores.ecmwf.int/cadl-arco-geo-011/arco/projections_cordex_domains_single_levels/eur11-rcp45-day-mohc_hadgem2_es-clmcom_clm_cclm4_8_17-r1i1p1/geoChunked.zarr"

# Time-chunked data for EUR-11, RCP 4.5, MOHC-HadGEM2-ES, CLMcom-CLM-CCLM4-8-17, r1i1p1, daily
timechunked_url = "https://arco.datastores.ecmwf.int/cadl-arco-time-011/arco/projections_cordex_domains_single_levels/eur11-rcp45-day-mohc_hadgem2_es-clmcom_clm_cclm4_8_17-r1i1p1/timeChunked.zarr"


# Use obstore's HTTPStore to create a store with retry configuration,
# and then wrap it in a zarr ObjectStore to read with xarray.
# See https://github.com/developmentseed/obstore/blob/main/obstore/python/obstore/_store/_retry.pyi
# for more details on the retry configuration options.
http_store = HTTPStore(
    geochunked_url,
    client_options={
        "default_headers": {"Authorization": f"Bearer {cdsapi_key}"},
    },
)
store = ObjectStore(http_store, read_only=True)
ds = xr.open_zarr(store)
ds
```

The result is the same as the previous example, this just provides a more robust connection, preventing your
workflow crashing due to temporary network issues.
